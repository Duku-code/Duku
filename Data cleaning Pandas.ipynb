```python
# Data Cleansing
'''
Remove Bad Data:
1. Empty Cells
2. Data in wrong format
3. Wrong Data
4. Duplicates
'''
```

```python
df = pd.read_csv('data.csv')
df
```

```python
import math
x = round(df['Calories'].mean(),1)
x
```

```python
y = round(df['Pulse'].mean(),1)
y
```

```python
df
```

```python
df['Calories'].fillna(x, inplace=True)
df
```

```python
df['Pulse'].fillna(y, inplace=True)
df
```

```python
import datetime
now = datetime.datetime.now()
date_time = now.strftime('%d-%m-%Y')
date_time
```

```python
df.at[27, 'Date'] = date_time
df
```

```python
# Removing Duplicates
df.drop_duplicates(inplace = True)
df
```

```python
df.info()
```

```python
# create new column
df['Tot_Pulse'] = df['Pulse'] + df['Maxpulse']
df
```

```python
# drop the column
df = df.drop(['Tot_Pulse'], axis='columns')
df
```

```python
# drop rows
df = df.drop(df.index[20:])
df
```

```python
df.info()
```

```python
df_cat = df.append(df_xlsx)
df_cat
```

```python
for row in df_cat.itertuples():
    if df_cat.at[row.Index, "Duration"] == 450:
        df_cat.at[row.Index, "Duration"] = 45
    elif df_cat.at[row.Index, "Duration"] == 600:
        df_cat.at[row.Index, "Duration"] = 60

df_cat
```

```python
customer=pd.DataFrame({
    'id':[1,2,3,4,5,6,7,8,9],
    'name':['Olivia','Aditya','Cory','Isabell','Dominic','Tyler','Samuel','Daniel','Jeremy'],
    'age':[20,25,15,10,30,65,35,18,23],
    'Product_ID':[101,0,106,0,103,104,0,0,107],
    'Purchased_Product':['Watch','NA','Oil','NA','Shoes','Smartphone','NA','NA','Laptop'],
    'City':['Mumbai','Delhi','Bangalore','Chennai','Chennai','Delhi','Kolkata','Delhi','Mumbai']
})
product=pd.DataFrame({
    'Product_ID':[101, 102, 103, 104, 105, 106, 107],
    'Product_name':['Watch', 'Bag', 'Shoes', 'Smartphone', 'Books', 'Oil', 'Laptop'],
    'Category':['Fashion', 'Fashion', 'Fashion', 'Electronics', 'Study', 'Grocery', 'Electronics'],
    'Price':[299.0, 1350.0, 2999.0, 14999.0, 145.0, 110.0, 79999.0],
    'Seller_City':['Delhi','Mumbai','Chennai','Kolkata','Delhi','Chennai','Bangalore']
})
```

```python
customer
```

```python
product
```

```python
pd.merge(product, customer, on='Product_ID')
```

```python
pd.merge(product, customer, left_on='Product_name', right_on='Purchased_Product')
```

```python
pd.merge(product,customer,how='inner',left_on=['Product_ID','Seller_City'],right_on=['Product_ID','City'])
```
