```python
import gradio as gr
from transformers import pipeline
```

```python
sentiment = pipeline('sentiment-analysis')
```

```python
def get_sentiment(input_text):
  return sentiment(input_text)
```

```python
result = get_sentiment("The movie is very bad")
```

```python
result
```

```python
iface = gr.Interface(fn = get_sentiment, inputs = "text", outputs = ['text'], title = 'Sentiment Analysis', description='Get sentiment Positive/Negative for a given input')
```

```python
iface.launch()
```
