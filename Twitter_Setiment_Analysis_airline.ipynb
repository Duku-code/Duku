```python
import pandas as pd
import numpy as np
import numpy as np
import seaborn as sns  
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/drive/MyDrive/Advanced AI/Tweets.csv')
```

```python
df.head()
```

```python
df_sliced = df[['text', 'airline_sentiment']]
```

```python
df_sliced['tidy_tweet'] = df_sliced['tidy_tweet'].replace(to_replace = '(@[\w]+)', value = '', regex=True)
df_sliced['tidy_tweet'] = df_sliced['tidy_tweet'].replace(to_replace = ('RT'), value = '', regex=True)
df_sliced['tidy_tweet'] = df_sliced['tidy_tweet'].replace(to_replace = ('\n'), value = '', regex=True)
```

```python
df_sliced.head()
```

```python
df_sliced["tidy_tweet"] = df_sliced["tidy_tweet"].astype(str).str.split()
```
