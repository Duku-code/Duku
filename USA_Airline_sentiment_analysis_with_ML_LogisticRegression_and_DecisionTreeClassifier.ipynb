```python
import pandas as pd
df= pd.read_csv('/content/Tweets.csv')
```

```python
df.head()
```

```python
df_sliced = df[['text','airline_sentiment']]
df_sliced.head()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# removing @mentions
df_sliced['tidy_tweet'] = df_sliced['text'].replace(to_replace = '(@[\w]+)', value = '', regex=True)
# REMOVE 'RT' 
df_sliced['tidy_tweet'] = df_sliced['tidy_tweet'].replace(to_replace = ('RT'), value = '', regex=True)
df_sliced['tidy_tweet'] = df_sliced['tidy_tweet'].replace(to_replace = ('\n'), value = '', regex=True)
```

```python
df_sliced.head()
```

```python
df_sliced["tidy_tweet"] = df_sliced["tidy_tweet"].str.split()
```

```python
df_sliced.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words("english")
```

```python
for i in df_sliced['tidy_tweet'][0:5]: 
  temp = [word for word in i if str(word).lower() not in stop]
  print(temp)
```

```python
df_sliced['tidy_tweet'][0][0][0:3]
```

```python
df_sliced['tidy_tweet'] = df_sliced['tidy_tweet'].apply(lambda x:[word for word in x if str(word).lower() not in stop])
```

```python
df_sliced
```

```python
# applying lemmatization on the text
import nltk
nltk.download('omw-1.4')
nltk.download('wordnet') # wordnet is a dictionary 
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer  = nltk.stem.WordNetLemmatizer()
def lemmatize_text(text):
  return [lemmatizer.lemmatize(str(w), pos ='v') for w in str(text).split()]
```

```python
def rejoin_words(row):
  mylist = row['tidy_tweet']
  joined_words = (" ".join(mylist))
  return (str(joined_words).lower())
df_sliced['tidy_tweet'] = df_sliced.apply(rejoin_words, axis = 1)
```

```python
df_sliced.head()
```

```python
df_sliced['tidy_tweet_lemma'] = df_sliced['tidy_tweet'].apply(lambda x: lemmatize_text(x))
```

```python
df_sliced['tidy_tweet_lemma']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_sliced['tidy_tweet'], df_sliced['airline_sentiment'], test_size = 0.2, random_state=2)
```

Passing the data in to CountVectorizer 

```python
cv = CountVectorizer(ngram_range=(1,1), 
                     stop_words = "english", 
                     min_df=3,  
                     max_df=0.1, 
                     max_features=1500)
x_train_cv = cv.fit_transform(x_train)
```

Change the y data in to  number using LabelEncoding 

```python
#using ITDF Veectorizer
tfdf = TfidfVectorizer(ngram_range=(1,1), 
                     stop_words = "english", 
                     min_df=3,  
                     max_df=0.1, 
                     max_features=1500)
x_train_tf = tfdf.fit_transform(x_train)
x_test_tf= tfdf.transform(x_test)
```

```python
df_check = pd.DataFrame(x_train_tf .toarray(), columns = cv.get_feature_names_out())
df_check
```

```python
#Apply labelEcoder to transform the y data to number 
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
y_train_le = le.fit_transform(y_train)
y_test_le = le.transform(y_test)
```

Apply Machine Learning modeel on the preprocessing 




```python
from sklearn.linear_model import LogisticRegression
LogReg= LogisticRegression()
LogReg.fit(x_train_tf, y_train_le)
```

```python
LogReg_pred = LogReg.predict(x_test_tf)
LogReg_pred[:4]
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test_le, LogReg_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train_tf, y_train_le)
```

```python
clf_pred = clf.predict(x_test_tf)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test_le, clf_pred ))
```
