```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue = 'tip', data=tips).set(title="Total Bill v/s Tip")
```

```python
sns.scatterplot(x='total_bill', y='tip', hue = 'sex', data=tips)
```

```python
sns.relplot(x='total_bill', y='size', hue = 'sex', data=tips)
```

```python
tips['sex'].value_counts()
```

```python
sns.relplot(x='total_bill', y='size', hue = 'smoker', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue = 'smoker', style='time', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue = 'smoker', kind='line', data=tips)
```

```python
# Categorical Scatterplot
```

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips)
```

```python
sns.catplot(x='smoker', y='tip', order=['No', 'Yes'], kind='violin', data=tips)
```

```python
df = sns.load_dataset("titanic")
sns.countplot(x=df["class"])
```

```python
sns.countplot(data=df, x="class", hue="alive")
```

```python
# Distribution Plot
```

```python
sns.displot(tips , x='total_bill', kind='kde', fill=True)
```

```python
sns.displot(tips , x='total_bill', hue='sex', kind='kde')
```

```python
sns.displot(tips , x='total_bill', hue='sex', element='step')
```

```python
# Pair Plot
```

```python
g = sns.pairplot(tips)
g.fig.set_size_inches(5,5)
```

```python
g = sns.pairplot(tips, hue='smoker')
g.fig.set_size_inches(5,5)
```

```python
sns.boxplot(x="total_bill", data=tips)
```

```python
sns.boxplot(x="total_bill", y='sex', data=tips)
```

```python
sns.boxplot(x="total_bill", y='smoker', data=tips)
```
